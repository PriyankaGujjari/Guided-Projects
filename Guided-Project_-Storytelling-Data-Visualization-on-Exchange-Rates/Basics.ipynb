## Cleaning the dataset

- As we are concerned only with comparing the euro and the US dollar currencies (scope of this project), we are going to rename only the releavant columns and leave the remaining columns intact.


- Convert the `Time` column to `datetime` data type.


- Sort the values of `Time` in ascending order.


- Reset the index and drop the initial index.

```python
exchange_rates.rename(columns = {'[US dollar ]': 'US_dollar', 
                                 'Period\\Unit:': 'Time'}, 
                      inplace = True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

exchange_rates.sort_values('Time', inplace = True)

exchange_rates.reset_index(drop = True, inplace = True)

exchange_rates.head(3)
```

```python
# Isolating the required columns into a new dataframe

euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

- If you observe , 62 observations have the value '-' for the column `US_dollar`. As such observations are of no use to use, it's best we eliminate the rows.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

- With this, we have cleaned the required data. The next step is to make visualizations.

```python
# Line plot to visualize the evolution of euro-dollar exchange rate

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(euro_to_dollar['Time'], 
         euro_to_dollar['US_dollar'])
plt.show()
```

- The graph above shows the daily variation of exchange rate over time. Thus, the graph is wiggly instead of a smooth line.


- But, we don't need the daily variation on our graph to show the long-term trends. To hide the daily variations, we can use **rolling mean** (also known as moving average)

```python
# Rolling mean for the `US_dollar` column

euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

## Affect of Brexit on the euro-dollar exchange rate

- Over the past 20 years, 2 major events impacted the Euro exchange rate.


    - The financial crash of 2008-09
    - The Brexit 2016-20
    
    
- The affect of the financial crisis is well-documented. So, in this project we are only going to deal with Brexit affect.


- Check the link for the affect of Brexit on Euro exchange rate.
[Link] (https://www.worldfirst.com/uk/blog/brexit/how-brexit-affects-pound-euro/)


- Note: We are only concerned about Euro in this project.

```python
brexit_pre = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2015) & 
                                   (euro_to_dollar['Time'].dt.year < 2016)]

brexit = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year < 2020)]

brexit_post = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2020)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

from datetime import datetime

# The plot
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(brexit_pre['Time'],brexit_pre['rolling_mean'],
        linewidth = 2, color = 'green')

ax.plot(brexit_post['Time'], brexit_post['rolling_mean'], 
        linewidth = 2, color = 'green')

# Highlighting the brexit timeframe (2016-20)

ax.plot(brexit['Time'], brexit['rolling_mean'], 
        linewidth = 3, color = 'red', label = 'Brexit timeframe')
# Highlighting the fall in exchange rate during brexit
ax.axvspan(xmin=(datetime(2016,6,1)), 
           xmax=(datetime(2017,3,16)), 
           ymin=0.02, alpha=0.3, color='grey')

# Title and Subtitles
ax.text((datetime(2015,1,1)),1.28, 
        "Affect of Brexit on euro-dollar rates",
        size= 24, weight ='bold')
ax.text((datetime(2015,1,1)), 1.25, 
        "(The shaded part shows the major downfall in euro-dollar rate)",
        size = 16)

### Adding a signature
ax.text((datetime(2014,1,1)), 1.00, '©DATAQUEST' + ' '*110 + 'Source: European Central Bank',
        color = 'white', backgroundcolor = 'black',
        size=10)
ax. legend(loc = 'upper right')
### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```

- From the graph above, it can be seen that the exchange rate decreased from June 2016 to March 2017 coinciding with the uncertainity in the events


    - Brexit Referendum (23rd June, 2016)
    - Referendum bill receives Royal Assent (16th March, 2017)
    
    
- Immediately after the 2-year period for the UK to leave EU started. During this period, UK funding to EU increased- thereby the euro exchange rate gained.


- UK was supposed to leave on 29th March,2019 but forwading the date. This created uncertainity and the markets responded again crashing the exchange rate.


- Post 2020, the exchange rate was majorly effected by rampant onset of **Covid-19 rather than Brexit**. So, that is beyond the scope of our project.
